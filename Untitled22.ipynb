```
# moving average on closing price - 21,7,9,50 and MACD indicator
df['ema_21'] = df['Close'].ewm(span=21, adjust=False).mean()
df['ema_7'] = df['Close'].ewm(span=7, adjust=False).mean()
df['ema_9'] = df['Close'].ewm(span=9, adjust=False).mean()
df['ema_50'] = df['Close'].ewm(span=50, adjust=False).mean()
df['macd'], df['macdsignal'], df['macdhist'] = MACD(df['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = scaler.fit_transform(df.drop(['Count'],axis=1))
```

```
##splitting dataset into train and test split
training_size=int(len(df)*0.65)
test_size=len(df)-training_size
train_data,test_data=df[0:training_size,:],df[training_size:len(df),:1]
```

```
training_size,test_size
```

```
import numpy
# convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return numpy.array(dataX), numpy.array(dataY)
```

```
# reshape into X=t,t+1,t+2,t+3 and Y=t+4
time_step = 100
X_train, y_train = create_dataset(train_data, time_step)
X_test, ytest = create_dataset(test_data, time_step)
```

```
from sklearn.ensemble import RandomForestRegressor
```

```
rf = RandomForestRegressor()
```

```
rf.fit(X_train,y_train)
```

```
y_pred = rf.predict(X_test)
```

```
from sklearn import metrics
print(metrics.mean_squared_error(ytest, y_pred))
print(metrics.mean_absolute_error(ytest, y_pred)) 
```
